# **TXT**

```python
from pyspark import SparkContext, SparkConf

if 'sc' in globals():
    sc.stop()

conf = SparkConf().setAppName("File").setMaster("local[*]")

sc = SparkContext(conf=conf)

rdd =sc.textFile("/content/sample_data/README.txt")

spark_lines = rdd.filter(lambda line: "mnist" in line)
print(spark_lines.collect())
```

# **CSV**

```python
from pyspark import SparkContext, SparkConf

rdd = sc.textFile("/content/test.csv")

spark_lines = rdd.map(lambda line: line.split(","))

filtered_rdd = spark_lines.filter(lambda cols: cols[1]=="Ariela" )

print(filtered_rdd.collect())

filtered_rdd.saveAsTextFile("/content")  # сохраняем RDD в текстовый файл по заданному пути

sc.stop()
```

# **Датафрейм**

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
.appName("read") \
.getOrCreate()

df = spark.read.text("/content/sample_data/README.txt")   # создаем датафрейм из текстового файла

df.show()

spark.stop()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
.appName("read") \
.getOrCreate()

# создаем датафрейм из csv файла
df = spark.read.option("header", "true").csv("/content/sample_data/mnist_test.csv")

df.show()

spark.stop()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
.appName("read") \
.getOrCreate()

data = [("Alice", 15), ("Bob", 25), ("Ivan", 35)]
columns = ["Name", "Age"]

df = spark.createDataFrame(data, columns)

# сохраняем датафрейм в csv  файл в папку
df.write.mode("overwrite").option("header", "true").csv("output")

df.show()

spark.stop()
```
